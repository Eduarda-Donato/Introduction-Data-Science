## Tratamento de Dados Faltantes

```python
import pandas as pd
```

```python
# NaN = valor nulo
dataset = pd.read_csv('dados/alugueis_residencial.csv', sep=';')
dataset.head(20)
```

```python
# A aplicação do método isnull() em um DataFrame retorna um DataFrame booleano, do mesmo tamanho do original, indicando se os valores são nulos.
dataset.isnull()
```

```python
#Contrario do isnull
# true para valores não NaN
dataset.notnull()
```

```python
#Mostra informações do df incluindo a quantidade de valores não nulos para cada variavel
dataset.info()
```

#### Eliminar todos os imoveis com valor NaN

```python
dataset[dataset['Valor'].isnull()]
```

```python
#Dropna elimina todos os dados com NaN em determinada variavel
qtd_imoveis_antes = dataset.shape[0]
dataset.dropna(subset= ['Valor'], inplace= True)
```

```python
qtd_imoveis_depois = dataset.shape[0]
print(qtd_imoveis_antes)
print(qtd_imoveis_depois)
print(f'QTD de imoveis com NaN no valor = {qtd_imoveis_antes-qtd_imoveis_depois}')
```

# -------------------------------

#### Eliminar todos os Apartamentos sem Condominios

```python
#quantidade de imoveis com NaN no condominio
dataset[dataset['Condominio'].isnull()].shape[0]
```

```python
qtd_imoveis_antes = dataset.shape[0]
selecao = (dataset['Tipo'] == 'Apartamento') & (dataset['Condominio'].isnull())
dataset = dataset[~selecao] #Para eliminar as assinaturas nulas selecionadas ao invés de coletá-las
dataset.info()
```

```python
qtd_imoveis_depois = dataset.shape[0]
print(qtd_imoveis_antes)
print(qtd_imoveis_depois)
print(f'QTD de imoveis com NaN no valor = {qtd_imoveis_antes-qtd_imoveis_depois}')
```

```python
dataset[dataset['Condominio'].isnull()].shape[0]
```

```python
dataset.fillna(0, inplace= True)
dataset.info()
```

#### Exportar 


```python
dataset.to_csv('dados/alugueis_residencial_tratado.csv', sep=';', index= False)
```
