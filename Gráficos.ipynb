# Gráficos

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize= (20,10))
```

```python
dataset = pd.read_csv('dados/alugueis_residencial_tratado_melhorado_sem_outliers.csv', sep=';')
dataset
```

```python
#Criar área para colocar os gráficos
area = plt.figure()
```

```python
#Inserir 4 graficos na área
#Dentro da figura tem 4 graficos, (2,2,x) 2 linhas. 2 colunas, posição do grafico
g1 = area.add_subplot(2,2,1)
g2 = area.add_subplot(2,2,2)
g3 = area.add_subplot(2,2,3)
g4 = area.add_subplot(2,2,4)

area
```

```python
#Criar um gráfico de dispersão (scatterplot)
g1.scatter(dataset.Valor, dataset.Area)
g1.set_title('Valor X Área')
area
```

```python
g2.hist(dataset.Valor)
g2.set_title('Histograma')
area
```

```python
#Criar uma amostra aleatória
amostra = dataset.Valor.sample(100).reset_index(drop=True)
amostra.head(20)
```

```python
g3.plot(amostra)
g3.set_title('Amostra (Valor)')
area
```

```python
grupo = dataset.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
```

```python
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
area
```
