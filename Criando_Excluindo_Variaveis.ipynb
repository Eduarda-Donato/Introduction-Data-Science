## Criando Novas Variáveis

### Variáveis:
### Valor Bruto;
### Valor de aluguel por m²;
### Valor Bruto por m²;
### Criar novo tipo para agregar os imoveis em casa ou ap.

```python
import pandas as pd
```

```python
dataset = pd.read_csv('dados/alugueis_residencial_tratado.csv', sep=';')
dataset
```

#### Valor Bruto

```python
dataset['Valor Bruto'] = (dataset['Valor'] + dataset['Condominio'] + dataset['IPTU'])
dataset
```

#### Valor por m²

```python
dataset['Valor m2'] = dataset['Valor'] / dataset['Area']
dataset
```

```python
dataset['Valor m2'] = dataset['Valor m2'].round(2)
dataset
```

#### Valor Bruto por m²

```python
dataset['Valor Bruto m2'] = (dataset['Valor Bruto'] / dataset['Area']).round(2)
dataset
```

#### Criar novo tipo para agregar os imoveis em casa ou ap

```python
#Para visualizar os tipos de imoveis
dataset['Tipo'].drop_duplicates()
```

```python
#Lista com a identificação das casas
casa = ['Quitinete', 'Casa', 'Casa de Condomínio', 'Casa de Vila']
#ap = ['Apartamento', 'Flat', 'Loft']
```

```python
#apply permite que execute uma função a cada linha/registro
dataset['Tipo Agregado'] = dataset['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dataset.head(30)
```
