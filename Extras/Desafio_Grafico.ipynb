```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(15,7))
```

```python
dataset = pd.read_csv('../dados/aluguel_amostra 2.csv',sep=';')
dataset
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(1,2,1)
g2 = area.add_subplot(1,2,2)
```

```python
dados = dataset['Tipo Agregado'].value_counts()
label = dataset['Tipo Agregado'].drop_duplicates()
g1.pie(dados, labels=label, autopct='%1.1f%%')
g1.set_title('Total de Imóveis por Tipo Agregado')
area
```

```python
grupo2 = dataset.groupby('Tipo')['Valor']
label = grupo2.count().index
valores = grupo2.count().values
g2.pie(valores, labels = label, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
g2.set_title('Total de Imóveis por Tipo')
area
```
